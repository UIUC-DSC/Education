## Installing packages

packages are libraries of code that we can "import" into our notebook and use by writing the package name along with the function name

`import pandas as pd`
`pd.read_csv`

when we import a package "as" something, we can use that name instead of the package name. but we could also write the following
`import pandas`
`pandas.read_csv`

pd is just shorter! a lot of packages have standard shortened names

pandas --> pd
matplotlib.pyplot --> plt
numpy --> np

sk-learn functions are usually imported individually:
`from sklearn.model_selection import train_test_split`
